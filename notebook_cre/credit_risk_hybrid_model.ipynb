```python
# --- Step 1: Import libraries ---
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split, StratifiedKFold, cross_val_score
from sklearn.metrics import (
    roc_auc_score, brier_score_loss, precision_score, recall_score,
    f1_score, accuracy_score, roc_curve, confusion_matrix
)
from scipy.stats.mstats import winsorize
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
from sklearn.feature_selection import SequentialFeatureSelector
from google.colab import files

# --- Step 2: Load your dataset ---
uploaded = files.upload()
file_name = list(uploaded.keys())[0]  # Получаем имя загруженного файла
df = pd.read_excel(file_name)

# Определение переменных
target = 'Default'
qualitative_vars = ['LI', 'AuditSt', 'ExtRtg']
quantitative_vars = [col for col in df.columns if col not in qualitative_vars + [target]]

# --- Step 4: Winsorization & Transformation ---
for col in quantitative_vars:
    df[col] = winsorize(df[col], limits=[0.01, 0.01])
    if df[col].skew() > 1:
        df[col] = np.log1p(df[col] - df[col].min() + 1)

# --- Step 4.3: Univariate Analysis – IV and p-values ---
def calc_iv(df, feature, target, bins=10):
    df_temp = df[[feature, target]].copy()
    df_temp['bin'] = pd.qcut(df_temp[feature].rank(method='first'), q=bins, duplicates='drop')
    grouped = df_temp.groupby('bin')[target].agg(['count', 'sum'])
    grouped['non_event'] = grouped['count'] - grouped['sum']
    grouped['event_rate'] = grouped['sum'] / grouped['sum'].sum()
    grouped['non_event_rate'] = grouped['non_event'] / grouped['non_event'].sum()
    grouped['woe'] = np.log((grouped['event_rate'] + 1e-6) / (grouped['non_event_rate'] + 1e-6))
    grouped['iv'] = (grouped['event_rate'] - grouped['non_event_rate']) * grouped['woe']
    return grouped['iv'].sum()

low_iv_vars = []
insignificant_vars = []

print("\n--- Univariate Analysis ---")
for col in quantitative_vars:
    try:
        iv = calc_iv(df, col, target)
        print(f"{col}: IV = {iv:.4f}")
        if iv < 0.02:
            low_iv_vars.append(col)
    except:
        print(f"{col}: IV calculation failed")

    try:
        X_uni = sm.add_constant(df[[col]])
        model = sm.Logit(df[target], X_uni).fit(disp=False)
        pval = model.pvalues[col]
        print(f"{col}: p-value = {pval:.4f}")
        if pval > 0.10:
            insignificant_vars.append(col)
    except:
        print(f"{col}: Univariate regression failed")

drop_candidates = list(set(low_iv_vars + insignificant_vars))
print(f"\nVariables marked for exclusion: {drop_candidates}")
quantitative_vars = [col for col in quantitative_vars if col not in drop_candidates]

# --- Step 4.4: Variance Inflation Factor (VIF) Filtering ---
from statsmodels.stats.outliers_influence import variance_inflation_factor

print("\n--- VIF Check for Multicollinearity ---")
X_vif = df[quantitative_vars].copy()
X_vif['Intercept'] = 1

vif_data = []
for i in range(X_vif.shape[1] - 1):  # exclude intercept
    vif = variance_inflation_factor(X_vif.values, i)
    vif_data.append((quantitative_vars[i], vif))
    print(f"{quantitative_vars[i]}: VIF = {vif:.2f}")

vif_threshold = 10
high_vif_vars = [var for var, vif in vif_data if vif > vif_threshold]

print(f"\nVariables with high VIF (> {vif_threshold}): {high_vif_vars}")
quantitative_vars = [col for col in quantitative_vars if col not in high_vif_vars]

# --- Step 5: Standardization ---
scaler = StandardScaler()
df[quantitative_vars] = scaler.fit_transform(df[quantitative_vars])

# --- Step 5.1: Exploratory Data Check ---
# Descriptive statistics
desc_stats = df.describe()

# Correlation matrix
correlation_matrix = df[quantitative_vars].corr()

# Identify highly correlated pairs (|r| > 0.9)
high_corr_pairs = []
threshold = 0.9
for i in range(len(correlation_matrix.columns)):
    for j in range(i):
        if abs(correlation_matrix.iloc[i, j]) > threshold:
            high_corr_pairs.append((
                correlation_matrix.columns[i],
                correlation_matrix.columns[j],
                correlation_matrix.iloc[i, j]
            ))

print("\nDescriptive statistics:\n", desc_stats)
print("\nHighly correlated pairs (|r| > 0.9):")
for var1, var2, corr_value in high_corr_pairs:
    print(f"{var1} vs {var2} | Correlation: {corr_value:.2f}")

# --- Step 6: Define datasets ---
X_quant = df[quantitative_vars]
X_full = df[quantitative_vars + qualitative_vars]
y = df[target]

Xq_train, Xq_test, y_train, y_test = train_test_split(X_quant, y, test_size=0.2, stratify=y, random_state=42)
Xf_train, Xf_test, _, _ = train_test_split(X_full, y, test_size=0.2, stratify=y, random_state=42)

print("\n--- Stepwise Feature Selection (Backward Elimination) ---")
logreg_base = LogisticRegression(max_iter=1000, solver='lbfgs')

# Perform backward stepwise selection using 5-fold CV and AUC as criterion
sfs = SequentialFeatureSelector(
    logreg_base,
    direction='backward',
    scoring='roc_auc',
    cv=5,
    n_jobs=-1
)
sfs.fit(Xf_train, y_train)

# Get selected feature names
selected_features = Xf_train.columns[sfs.get_support()]
print(f"Selected features ({len(selected_features)}): {list(selected_features)}")

# Filter training and test sets accordingly
Xf_train = Xf_train[selected_features]
Xf_test = Xf_test[selected_features]

# --- Step 7: Define models ---
models = {
    'Logistic Regression': LogisticRegression(max_iter=1000),
    'Random Forest': RandomForestClassifier(n_estimators=100, random_state=42),
    'Gradient Boosting': GradientBoostingClassifier(n_estimators=100, learning_rate=0.1, max_depth=3, random_state=42),
    'SVM (RBF Kernel)': SVC(kernel='rbf', probability=True, random_state=42)
}

# --- Step 8: Evaluation function ---
def evaluate_models(X_train, X_test, name_suffix):
    results = []
    for name, model in models.items():
        model.fit(X_train, y_train)
        y_pred_proba = model.predict_proba(X_test)[:, 1]
        y_pred = (y_pred_proba >= 0.5).astype(int)
        results.append({
            'Model': f"{name} {name_suffix}",
            'AUC': roc_auc_score(y_test, y_pred_proba),
            'Brier Score': brier_score_loss(y_test, y_pred_proba),
            'Accuracy': accuracy_score(y_test, y_pred),
            'Precision': precision_score(y_test, y_pred),
            'Recall': recall_score(y_test, y_pred),
            'F1 Score': f1_score(y_test, y_pred)
        })
    return results

# --- Step 9: Run model comparisons ---
results_quant = evaluate_models(Xq_train, Xq_test, "(Quant)")
results_full = evaluate_models(Xf_train, Xf_test, "(Quant+Qual)")
all_results = pd.DataFrame(results_quant + results_full)

# --- Step 10: Plot AUC Comparison ---
plt.figure(figsize=(12,6))
sns.barplot(data=all_results, x='Model', y='AUC', palette='viridis')
plt.xticks(rotation=45, ha='right')
plt.title('AUC Comparison Across Models and Feature Sets')
plt.tight_layout()
plt.grid(True)
plt.show()

# --- Step 11: Show Results Table ---
print(all_results)

# --- Step 12: Logistic Regression Final Model with 5-Fold CV ---
final_model = LogisticRegression(penalty='l2', solver='lbfgs', max_iter=1000)
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
cv_auc_scores = cross_val_score(final_model, Xf_train, y_train, cv=cv, scoring='roc_auc')
print(f"\n5-Fold CV AUC (LogReg with Quant+Qual): {np.mean(cv_auc_scores):.4f}")

# --- Step 13: Fit Final Model on Full Training Set ---
final_model.fit(Xf_train, y_train)
y_pred_proba = final_model.predict_proba(Xf_test)[:, 1]
y_pred = (y_pred_proba >= 0.5).astype(int)

# --- Step 14: Final Model Metrics ---
print(f"\nFinal Logistic Regression Model Evaluation on Test Set:")
print(f"AUC: {roc_auc_score(y_test, y_pred_proba):.4f}")
print(f"Brier Score: {brier_score_loss(y_test, y_pred_proba):.4f}")
print(f"Accuracy: {accuracy_score(y_test, y_pred):.4f}")
print(f"Precision: {precision_score(y_test, y_pred):.4f}")
print(f"Recall: {recall_score(y_test, y_pred):.4f}")
print(f"F1 Score: {f1_score(y_test, y_pred):.4f}")
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("""\n5-Fold CV AUC (LogReg with Quant+Qual): 0.9174\n\nFinal Logistic Regression Model Evaluation on Test Set:\nAUC: 0.9144\nBrier Score: 0.1156\nAccuracy: 0.8250\nPrecision: 0.8095\nRecall: 0.8500\nF1 Score: 0.8293\nConfusion Matrix:\n [[32  8]\n [ 6 34]]""")

rating_df = pd.DataFrame({
    'Actual_Default': y_test.values,
    'Predicted_PD': y_pred_proba
})

# Separate known defaults (rating 6) and non-defaults
non_defaults = rating_df[rating_df['Actual_Default'] == 0].copy()
defaults = rating_df[rating_df['Actual_Default'] == 1].copy()

# Step 1: PD quantiles for non-defaults (Q20 to Q80)
quantiles = non_defaults['Predicted_PD'].quantile([0.2, 0.4, 0.6, 0.8])
q20, q40, q60, q80 = quantiles[0.2], quantiles[0.4], quantiles[0.6], quantiles[0.8]

print("\n--- PD Quantile Thresholds (Non-Defaults) ---")
print(f"Q20: {q20:.4f}, Q40: {q40:.4f}, Q60: {q60:.4f}, Q80: {q80:.4f}")

# Step 2: Assign ratings 1–5 based on PD ranges
def assign_rating(pd_value):
    if pd_value <= q20:
        return 1
    elif pd_value <= q40:
        return 2
    elif pd_value <= q60:
        return 3
    elif pd_value <= q80:
        return 4
    else:
        return 5

non_defaults['Rating'] = non_defaults['Predicted_PD'].apply(assign_rating)
defaults['Rating'] = 6  # Step 3: Defaults are assigned Rating 6

# Step 4: Combine both and analyze results
rated_df = pd.concat([non_defaults, defaults]).sort_index()

print("\n--- Rating Distribution ---")
print(rated_df['Rating'].value_counts().sort_index())

# Step 5: Check actual default rates by rating
rated_df['Bucket'] = rated_df['Rating']
bucket_stats = rated_df.groupby('Bucket')['Actual_Default'].agg(['count', 'sum'])
bucket_stats['Default_Rate'] = bucket_stats['sum'] / bucket_stats['count']
print("\n--- Default Rate by Rating Category ---")
print(bucket_stats)

# --- Step 15: ROC Curve ---
fpr, tpr, _ = roc_curve(y_test, y_pred_proba)
plt.figure(figsize=(8,6))
plt.plot(fpr, tpr, label=f'ROC Curve (AUC = {roc_auc_score(y_test, y_pred_proba):.2f})')
plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve - Final Logistic Regression')
plt.legend(loc='lower right')
plt.grid(True)
plt.show()
```
